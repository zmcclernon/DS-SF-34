```python
df2 = df[ ['FATALITIES', 'EVTYPE', 'INJURIES', 'PROPDMG', 'CROPDMG', 'PROPDMGEXP', 'CROPDMGEXP', 'STATE', 'BGN_DATE']]

if df2['PROPDMGEXP'] != 'K' or df2['PROPDMGEXP'] != 'M' or df2['PROPDMGEXP'] != 'B':
    df2 = df2.drop(df2['PROPDMGEXP'], axis = 0)
elif df2['PROPDMGEXP'] == 'K':
    df2['PROPDMG'] == df2['PROPDMG'] * 1000
elif df2['PROPDMGEXP'] == 'M':
    df2['PROPDMG'] == df2['PROPDMG'] * 1000000
else:
    df2['PROPDMG'] == df2['PROPDMG'] * 1000000000
    
if df2['CROPDMGEXP'] != 'K' or df2['CROPDMGEXP'] != 'M' or df2['CROPDMGEXP'] != 'B':
    False
elif df2['CROPDMGEXP'] == 'K':
    df2['CROPDMG'] == df2['CROPDMG'] * 1000
elif df2['CROPDMGEXP'] == 'M':
    df2['CROPDMG'] == df2['CROPDMG'] * 1000000
else:
    df2['CROPDMG'] == df2['CROPDMG'] * 1000000000
    

df2['TOTALCOST'] == df2['PROPDMG'] + df2['CROPDMG']
```

```python
df2

df2['ADJPROPDMG'] = df2.PROPDMGEXP.map({'K' : 1000, 'M' : 1000000, 'B' : 1000000000}) * df2.PROPDMG
df2['ADJPROPDMG']
```

```python
df2['ADJCROPDMG'] = df2.CROPDMGEXP.map({'K' : 1000, 'M' : 1000000, 'B' : 1000000000}) * df2.CROPDMG
df2['ADJCROPDMG']
```
